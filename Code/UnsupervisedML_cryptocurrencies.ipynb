```python
# RECOMMENDATION: The output from t-sne and K-means clustering displayed that crypto currencies cannot be clustered together in any number of clusters.That being said, distinct groups of currencies were not identified and, therefore, no trends coudl be drawn.
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Read `crypto_data.csv` into Pandas.
```

```python
cryptocurrency_df = pd.read_csv("Data/crypto_data.csv")
```

```python
cryptocurrency_df
```

```python
# Discard all cryptocurrencies that are not being traded.
```

```python
cryptocurrency_df.info()
```

```python
cryptocurrency_df2 = pd.DataFrame(cryptocurrency_df)
```

```python
cryptocurrency_df2.query('IsTrading == True & TotalCoinsMined > 0', inplace=True)
cryptocurrency_df2
```

```python
cryptocurrency_df2.info()
```

```python
tradedcrypto_df = cryptocurrency_df2.drop(['Unnamed: 0', 'IsTrading', 'CoinName'], axis=1)

tradedcrypto_df
```

```python
#Converting "totalCoinsMined" and "TotalCoinSupply" to numeric
```

```python
tradedcrypto_df['TotalCoinSupply'] = tradedcrypto_df['TotalCoinSupply'].astype(float)

tradedcrypto_df
```

```python
#Create dummy variables to convert string values to numeric
```

```python
cryptofinal = pd.get_dummies(tradedcrypto_df)

cryptofinal
```

```python
#The "get dummies" convertion expanded the database into 98 columns.
```

```python
#Standardize the dataset  by scaling the dataset.
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(cryptofinal)
```

```python
scaled_crypto = pd.DataFrame(scaled_df, columns=cryptofinal.columns[0:])
scaled_crypto

scaled_crypto
```

```python
#Use Principal Component Analysis for dimensionality reduction.
```

```python
pca = PCA(n_components=0.90)
```

```python
cryptoreduced = pca.fit_transform(scaled_crypto)
```

```python
cryptoreduced_df = pd.DataFrame(data=cryptoreduced)
cryptoreduced_df.head()
```

```python
# After using PCA and preserving 90% of the explained variance in dimensionality reduction, the number of features were reduced from 98 to 74.
```

```python
pca.explained_variance_ratio_
```

```python
# Using t-SNE for further dimentionality reduction.
```

```python
t_sne = TSNE(learning_rate=300)
tsnereduced = t_sne.fit_transform(cryptoreduced)
```

```python
tsnereduced.shape
```

```python
# Using a scatterplot to display results
```

```python
plt.scatter(tsnereduced[:,0], tsnereduced[:,1])
plt.show()
```

```python
# For-loop to determine the inertia for each k, 1 through 10.
```

```python
inertia = []
k = list(range(1, 11))
```

```python
for i in k:
    k_means = KMeans(n_clusters=i, random_state=0)
    k_means.fit(cryptoreduced_df)
    inertia.append(k_means.inertia_)
```

```python
# Use hvPlot for Elbow Curve.
```

```python
elbowcurve = {"k": k, "inertia": inertia}
elbowcurve_df = pd.DataFrame(elbowcurve)

elbowcurve_df.head()
```

```python
# Elbow Curve plot.
```

```python
plt.plot(elbowcurve_df['k'], elbowcurve_df['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
